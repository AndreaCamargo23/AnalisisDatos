```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('penguins_size.csv')

df.boxplot(column='culmen_length_mm')
plt.title('Boxplot de culmen_length_mm')
plt.ylabel('Longitud (mm)')
plt.grid(True)
plt.show()
condicion_sin_sexo = ~df["sex"].isin(['MALE', 'FEMALE'])
df.loc[condicion_sin_sexo, 'sex'] = 'UNKNOWN'
print("Pinguinos sin Sexo", df[~df["sex"].isin(['MALE', 'FEMALE'])])
```
